```python
#  Q.11 How does the merge() function in Pandas differ from the concat() function?
# both used for combining datasets or dataframes
# concat add data after 1 dataframe

import pandas as pd
df1 = pd.DataFrame({'id': [1, 2, 3], 'name': ['asd', 'Bob', 'aadi']})
df2 = pd.DataFrame({'id': [2, 3, 4], 'score': [85, 90, 75]})
result = pd.merge(df1, df2, on='id')
print(result)
```

```python
result = pd.concat([df1, df2])
result
```

```python
# Q.12 What is the purpose of the pivot_table() function in Pandas?

import pandas as pd
data = {'Product': ['A', 'B', 'A'], 'Region': ['East', 'West', 'East'], 'Sales': [100, 200, 150]}
df = pd.DataFrame(data)
pivot = pd.pivot_table(df, values='Sales' , columns='Region')
print(pivot)
```

```python
# Q.13 Explain the difference between loc[] and iloc[] in Pandas.
# used to select subsets of data from a DataFrame or Series
# loc[] (Label-based Indexing)
# iloc[] (Integer-based Indexing)

import pandas as pd
data = {'A': [10, 20, 30], 'B': [40, 50, 60], 'C': [70, 80, 90]}
df = pd.DataFrame(data)

# Using loc[]
print(df.loc[0,'A'])

# Using iloc[]
print(df.iloc[0, 0])
```
