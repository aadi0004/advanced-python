```python
#Q.1 How do you create a Pandas DataFrame?
import pandas as pd
data= {
    'Name':['Alice','Bob','Claire'],
    'Age':[25,30,27],
    'Gender':['Female','Male','Female']
}
df=pd.DataFrame(data)
print(df)
```

```python
# Q.2 How do you read a CSV file into a Pandas DataFrame?
import pandas as pd
df=pd.read_csv('covid_toy.csv')
df.head(5)
```

```python
# Q.3 How do you check for missing values in a DataFrame?
df.isnull().sum()
```

```python
# Q.4 How do you drop rows with missing values?
df.dropna()
```

```python
#Q.5 How do you fill missing values with a specific value?
df.fillna(0)
```

```python
# Q.6 How do you filter rows based on a condition?
df[df['age']>20]
```

```python
# Q.7 How do you select a specific column in a DataFrame?
df['age']
```

```python
# Q.8 How do you rename columns in a DataFrame?
df.rename(columns={'has_covid':'has_Covid'})
```

```python
# Q.9 How do you sort a DataFrame by a specific column?
df.sort_values(by= 'age')
```

```python
# Q.10 How do you group data and calculate aggregate statistics?
df.groupby('has_covid').sum()
```

```python
# Q.11 How do you reset the index of a DataFrame?
# df.reset_index()
df.reset_index(drop=True, inplace=True)
```

```python
# Q.12 How do you merge two DataFrames?
import pandas as pd
df1=pd.DataFrame({'id':[12,34],'name':['asd','qwe']})
df2=pd.DataFrame({'id':[12,34,45],'name':['ram','shu','df']})
pd.merge(df1,df2,on='id')
```

```python
# Q.13 How do you concatenate two DataFrames?
pd.concat([df1,df2])
```

```python
# Q.14 How do you pivot a DataFrame?
df.pivot(columns='has_covid',values='age')
```

```python
# Q.15 How do you melt a DataFrame?
df.melt(id_vars='has_covid',value_vars='age')
```
